### Creating a helper function for getting basic info out of dataset

```python
def summary(df):
    print(f"Dataset has {df.shape[1]} features and {df.shape[0]} examples.")
    summary = pd.DataFrame(index=df.columns)
    summary["Unique"] = df.nunique().values
    summary["Missing"] = df.isnull().sum().values
    summary["Duplicated"] = df.duplicated().sum()
    summary["Types"] = df.dtypes
    return summary

summary(df)
```

<h2 style="color:rgb(228, 33, 12); text-align:center; font-weigth: bold;"> 👨‍🚒Feature Engineering🏗️</h2>

**Note: we will be applying feature engineering directly on the concatenated dataset**
Given Information about the dataset:
* `PassengerId` - A unique Id for each passenger. Each Id takes the form gggg_pp where gggg indicates a group the passenger is travelling with and pp is their number within the group. People in a group are often family members, but not always.
* `HomePlanet` - The planet the passenger departed from, typically their planet of permanent residence.
* `CryoSleep` - Indicates whether the passenger elected to be put into suspended animation for the duration of the voyage. Passengers in cryosleep are confined to their cabins.
* `Cabin` - The cabin number where the passenger is staying. Takes the form deck/num/side, where side can be either P for Port or S for Starboard.
* `Destination` - The planet the passenger will be debarking to.
* `Age` - The age of the passenger.
* `VIP` - Whether the passenger has paid for special VIP service during the voyage.
* `RoomService`, `FoodCourt`, `ShoppingMall`, `Spa`, `VRDeck` - Amount the passenger has billed at each of the Spaceship Titanic's many luxury amenities.
* `Name` - The first and last names of the passenger.
* `Transported` - Whether the passenger was transported to another dimension. This is the target, the column you are trying to predict.

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'mean')
```

### PassangerID
`PassangerId` is composed of group id and pp represents the person's number in the group itself. We can transform this information into new feature as `groupsize` that tell us the size of the group of the person. It will be more helpful.

```python
groups = df['PassengerId'].apply(lambda x: x.split('_')[0]).value_counts().to_dict()
df['group_size'] = df['PassengerId'].apply(lambda x: groups[x.split('_')[0]])
df.set_index('PassengerId', inplace=True)
```

### HomePlanet

```python
temp = df['HomePlanet'].value_counts()
temp
```

```python
p = temp.values 
p = p/sum(p)
p
```

```python
size = df['HomePlanet'].isna().sum()
```

```python
df.loc[df['HomePlanet'].isna(),'HomePlanet'] = np.random.choice(['Europa', 'Earth', 'Mars'],size, p=p)
```

### CryoSleep
We will be filling `nan` values present in this column based on the fact that if a passenger's cryosleep status is not recorded that means, he probably not in CryoSleep.

```python
df['CryoSleep'].value_counts()
```

```python
df['CryoSleep'].fillna(value = False, inplace = True)
```

```python
df.loc[:,'CryoSleep'] = df.apply(lambda x: False if x.luxury != 0 else x,axis =1)
```

```python
df['CryoSleep'] = df.CryoSleep.astype(int)
```

### Destination
We will follow the same approach as of we follow in `HomePlanet`

```python
temp = df['Destination'].value_counts()
dest = temp.index
prob = temp.values/sum(temp.values)
size = df['Destination'].isna().sum()
```

```python
df.loc[df['Destination'].isna(),'Destination'] = np.random.choice(dest,size, p=prob)
```

### VIP
* Total `nan` values count: 296.
* We will be filling `nan` values present in this column with `False` based on the fact that if a passenger's VIP status is not recorded that means, he probably not aVIP.

```python
df['VIP'].fillna(value = False, inplace = True)
df['VIP'] = df['VIP'].astype(int)
```

```python
df['VIP'].isna().sum()
```

### Cabin
Total number null values: 299.
Aparting from filling those nan values, we can also extract the useful information from the Cabin and those are `Deck`, `Cabin num`, `Cabin side`.For understanding purpose, we can understand it by analogy (building number -- deck, wing -- number, flat -- side).

```python
df[['Deck', 'Cabin_num','Cabin_side']] = df['Cabin'].str.split('/', expand = True,)
```

Let's perform feature Engineering on each of them one by one:
#### Cabin-Deck
One important thing to observe that more than `70%` of the passangers are travelling only in top 2 Deck i.e. F and G. It is very helpful in feeling the missing value in `Deck` column. Instead of using strategy of most-frequent, we should fill the nan values with those top 2 cabin decks.

```python
df['Deck'].value_counts()
```

```python
df.loc[df['Deck'].isna(), 'Deck'] = np.random.choice(['F', 'G'],df['Deck'].isna().sum(),p=[0.53, 0.47])
```

```python
df['Deck'].isna().sum()
```

#### Cabin_num
In it we will use Simple imputer technique with strategy mean.

```python
df['Cabin_num'] = imputer.fit_transform(df['Cabin_num'].values.reshape(-1,1))
```

#### Cabin_side
In this we simply fill `nan` values randomly with the two available category and then convert them into numbers by assigning `P` to 1 and `S` to 0.

```python
df.loc[df['Cabin_side'].isna(), 'Cabin_side'] = np.random.choice(['P','S'],df['Cabin_side'].isna().sum(),
                                                                p = [0.5,0.5])
```

```python
df['Cabin_side'] = np.where(df['Cabin_side']=='P',1,0)
```

### Age
Total number of missing values: 270
To handle these missing values, we will fill them with a uniform distribution within one standard deviation of the mean age.

```python
mean_age = df['Age'].mean()
age_std = df['Age'].std()
upper = mean_age + age_std
lower = mean_age-age_std
df.loc[df['Age'].isna(),'Age'] = np.random.uniform(upper, lower, size = df.Age.isna().sum())
```

### RoomService, FoodCourt, ShoppingMall, Spa, VRDeck
* We have different number of values in each of the column, to fill them we can't use mean as it is not robust to outliers, instead of it, we will use strategy as median.
* Then we create a new column as `luxury` which tells about the total spending of the passanger in one column.
* To deal with outliers, we will log scale the data and log(0) is not defined, hence we will put 0.235(any small values you can put) at that place. This helps to preserve the original distribution of the data.

```python
from sklearn.impute import SimpleImputer
median_imputer = SimpleImputer(strategy = 'median')
```

```python
for col in luxury_cols:
    df[col] = median_imputer.fit_transform(df[col].values.reshape(-1,1))
```

```python
luxury_cols.append('luxury')
```

```python
fig, axes = plt.subplots(len(luxury_cols),2, figsize=(12,14))
for i, col in enumerate(luxury_cols):
    sns.histplot(data=df, x=col, ax=axes[i, 0], bins=20)
    sns.histplot(data=np.log(df[[col]]), x=col, ax=axes[i, 1])
    axes[i, 0].set_title('Normal Distribution')
    axes[i, 1].set_title('Logarithmic Distribution')
plt.tight_layout()
```

```python
for col in luxury_cols:
    df.loc[df[col]==0, col] = 0.356
    df[col] = np.log(df[col])
```

## Hypertuning LightGBM

```python
from sklearn.model_selection import RandomizedSearchCV

# Define the parameter grid to search
param_grid = {
    'n_estimators': [300, 500,700,900,1100],
    'learning_rate': [0.01, 0.1, 0.3],
    'max_depth': [3, 5, 7],
    'num_leaves': [20, 30, 40],
    'reg_alpha': [0, 0.1, 0.5],
    'reg_lambda': [0, 0.1, 0.5]
}

# Create the LGBMClassifier
lgbm = LGBMClassifier()

# Perform random search with cross-validation
random_search = RandomizedSearchCV(estimator=lgbm, param_distributions=param_grid, n_iter=10, cv=5)

# Fit the random search to the training data
random_search.fit(X_train, y_train)

# Get the best hyperparameters and the best score
best_params = random_search.best_params_
best_score = random_search.best_score_

# Create a new LGBMClassifier with the best hyperparameters
best_lgbm = LGBMClassifier(**best_params)

# Train the model using the best hyperparameters
best_lgbm.fit(X_train, y_train)
```

```python
best_params, best_score
```

```python
best_lgb_params = {'reg_lambda': 0.5,
  'reg_alpha': 0.1,
  'num_leaves': 30,
  'n_estimators': 500,
  'max_depth': 3,
  'learning_rate': 0.1}
```

```python
y_pred = best_lgbm.predict(X_val)
acc_lgm = accuracy_score(y_val,y_pred)
scores.append(['Only best lgm', acc_lgm])
print("Accuracy",acc_lgm)
```
